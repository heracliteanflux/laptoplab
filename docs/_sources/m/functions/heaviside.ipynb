---

__Heaviside Step Function__ [Wiki](https://en.wikipedia.org/wiki/Heaviside_step_function)<br>
$$
H(x)=
\begin{cases}
1 & x \gt 0 \\
0 & x \lt 0 \\
\end{cases}
$$
