# Ramp Function

---

__Ramp Function__ [Wiki](https://en.wikipedia.org/wiki/Ramp_function)<br>

$
\begin{aligned}
R(x)=
\begin{cases}
x & x \ge 0 \\
0 & x \lt 0 \\
\end{cases}
=\text{max}\, (x, 0)
\end{aligned}
$
