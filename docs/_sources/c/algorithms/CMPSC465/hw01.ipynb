De-duplication is a common data processing task: taking an array containing values and removing all the duplicate elements. There exist a wide range of different algorithms for completing this task efficiently.

Consider an unsorted array $A$ that contains $n$ values which supports three operations:
1. `get(A, i)` $= O(1)$ returns the value of the $i$-th element
2. `search(A, i)` $= O(n)$ searches the array $A$ for all elements at indexes larger than $i$ that are the same as the element at index $i$ using a linear search, and returns the indexes as a list
3. `delete(A, i)` $= O(n)$ deletes the element at index $i$ from the array by shifting all elements at index $i + 1$ to $n - 1$ down by one

You may assume that the binary comparison operation is $O(1)$ and that the maximum size of the output array is a small constant and so the space and time costs of handling it can be neglected.

Propose an algorithm for performing de-duplication of an array of elements in-place, with $O(n^3)$ worst-case time complexity and $O(1)$ space complexity. Prove both the correctness of this algorithm and its time and space complexity bounds.

---

```python
A = array
k = 0

while (k < len(A)):
  list_of_indexes = search(A, k)   # O(n)
  m = len(list_of_indices) - 1     # O(1)
  while (m >= 0):
    delete(A, list_of_indexes[m])  # O(n)
    m = m - 1                      # O(1)
  k = k + 1
```

When our array contains $n$ copies of a single value as in `1 1 1 1 1`, the outer while loop iterates $1$ time and the inner while loop iterates $n - 1$ times per outer loop iteration.

$T(n) = \underbrace{(\textcolor{red}{1})(\textcolor{orange}{n} + \textcolor{yellow}{c_1})}_{\text{outer}}\underbrace{(\textcolor{green}{n - 1})(\textcolor{blue}{n} + \textcolor{purple}{c_2})}_{\text{inner}}$
* <span style="color: red;">outer loop</span>
* <span style="color: orange;">search(A, i)</span>
* <span style="color: yellow;">m = len(list_of_indexes) - 1, i = i + 1</span>
* <span style="color: green;">inner loop</span>
* <span style="color: blue;">delete(A, i)</span>
* <span style="color: purple;">m = m + 1</span>

When our array contains no duplicate values as in `1 2 3 4 5`, the outer loop iterates $n - 1$ times and the inner loop iterates $1$ time per outer loop iteration.

$T(n) = \underbrace{(\textcolor{red}{n - 1})(\textcolor{orange}{n} + \textcolor{yellow}{c_1})}_{\text{outer}}\underbrace{(\textcolor{green}{1})(\textcolor{blue}{n} + \textcolor{purple}{c_2})}_{\text{inner}}$
* <span style="color: red;">outer loop</span>
* <span style="color: orange;">search(A, i)</span>
* <span style="color: yellow;">m = len(list_of_indexes) - 1, i = i + 1</span>
* <span style="color: green;">inner loop</span>
* <span style="color: blue;">delete(A, i)</span>
* <span style="color: purple;">m = m + 1</span>

In either case, the equation yields the same result because of the commutativity of integer factors in a product.

$
\begin{aligned}
T(n)
&= (1)(n + c_1)(n - 1)(n + c_2) \\
&= (n^2 - n + c_1n - c_1)(n + c_2) \\
&= n^3 + c_2n^2 - n^2 - c_2n + c_1n^2 + c_1c_2n - c_1n - c_1c_2 \\
&= n^3 + c_1n^2 + c_2n^2 - n^2 + c_1c_2n - c_1n - c_2n - c_1c_2 \\
&= n^3 + (c_1 + c_2 - 1)n^2 + (c_1c_2 - c_1 - c_2)n - c_1c_2 \\
&= n^3 + an^2 + bn - c \\
&= O(n^3)
\end  {aligned}
$

Pre-condition
* $k = 0$
* $n$ is a positive integer
* The array $A = a_0, \dots, a_{n - 1}$ consists of $n$ (possibly duplicated) values
* $k \lt n = |A|$

Loop Condition $C$
* $k \lt |A|$

Loop Invariant $I$
* $k \le |A| \le n$
* The first $k$ elements of the array $A = a_0, \dots, a_{k - 1}$ consist of non duplicated values

Post-condition
* $k = |A| \le n$
* The array $A = a_0, \dots, a_{k - 1}$ consists of non duplicated values.

Assume the pre-condition, that $k = 0$, that $n$ is a positive integer, and that the array $A = a_0, \dots, a_{n - 1}$ consists of $n$ (possibly duplicated) values.

Then it is true that $I$ holds:
* $k \lt n = |A|$ since $k = 0$ and $n \ge 1$ and this implies $k \le |A| \le n$
* the first $k = 0$ elements of the array consist of non duplicated values

STEP 2 - show that $C, I_1 \implies I_2$

Assume that $C, I$ hold pre iteration:
* $k_1 \lt |A|$
* $k_1 \le |A| \le n$
* the first $k_1$ elements of the array consist of non duplicated values

Then it is true that $I$ holds post iteration:
* $k_2 = k_1 + 1$
* $k_1 \lt |A| \implies k_1 + 1 \le |A|$
* the first $k_1 + 1$ elements of the array consist of non duplicated values

STEP 3 - Eventually it will be the case that $\lnot C$ after a finite number of steps.

After $n$ iterations $k = n = |A|$

STEP 4

Assume that $\lnot C, I$
* $k \ge |A|$
* $k \le |A| \le n$
* $\implies k = |A| \le n$
