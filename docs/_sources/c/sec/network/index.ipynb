
---

## nmap

[ [d](https://nmap.org/ndiff/) ] ndiff

* [ [y](https://www.youtube.com/watch?v=JHAMj2vN2oU) ] `02-10-2024`. Hacker Joe. "NMAP Full Guide (You will never ask about NMAP again)".

---

## Virtual Private Network

* [ h ] OpenVPN
* [ [h](https://www.pivpn.io) ] PiVPN
* [ [h](https://www.wireguard.com) ] Wireguard

---

## Resources

Tools & Technologies
* [ [h](https://flowgrind.github.io/) ][ [w](https://en.wikipedia.org/wiki/Flowgrind) ] Flowgrind
* [ [h](https://iperf.fr/) ][ [w](https://en.wikipedia.org/wiki/Iperf) ] iperf - TCP/UDP bandwidth measurement
* [ [h](https://hewlettpackard.github.io/netperf/) ][ [w](https://en.wikipedia.org/wiki/Netperf) ] netperf - network performance benchmark
* [ [h](https://www.nsnam.org) ][ w ] ns-3 - modeling and simulation
* [ [h](https://www.ntop.org/) ][ [w](https://en.wikipedia.org/wiki/Ntop) ] ntopng
  * https://www.ntop.org/get-started/download/
  * https://www.ntop.org/guides/ntopng/what_is_ntopng.html#installing-on-macos
  * https://www.ntop.org/products/traffic-analysis/ntop/
* [ [h](https://www.pfsense.org/) ][ d ][ g ][ w ] pfSense
* [ [h](https://scapy.net/) ][ d ][ g ][ w ] Scapy - packet manipulation
* [ [h](https://www.wireshark.org/) ][ d ][ [g](https://gitlab.com/wireshark/wireshark) ][ [w](https://en.wikipedia.org/wiki/Wireshark) ] Wireshark - network traffic capture and analysis

Courses
* [ [h](https://online.stanford.edu/courses/cs144-introduction-computer-networking) ] Stanford CS144 Introduction to Computer Networking

Online
* [ [h](https://www.submarinecablemap.com/) ] Submarine Cable Map
* https://www.redhat.com/sysadmin/test-tcp-python-scapy

border gateway protocol
* [ [y](https://www.youtube.com/watch?v=ZBvplOms19U) ] `08-04-2021`. Kevin Du. "BGP Routing and Attacks".
* [ [y](https://www.youtube.com/watch?v=-wMU8vmfaYo) ] `10-12-2021`. Ben Eater. "Why was Facebook down for five hours?".
* [ [y](https://www.youtube.com/watch?v=2pJGOF8EHQ8) ] `12-04-2020`. Epic Networks Lab. "BGP Hijacking - Detection and Prevention | Network Traffic Analysis Ep. 16 | CS4558".
* [ [y](https://www.youtube.com/watch?v=39DGVpMt7eQ) ] `09-20-2021`. David Bombal. "Destroying the Internet (BGP routers) EP 1 // BGP Python scapy DoS script".
* [ [y](https://www.youtube.com/watch?v=O6tCoD5c_U0) ] `05-04-2016`. Computerphile. "BGP: Border Gateway Protocol - Computerphile".

David Bombal
* [ [y](https://www.youtube.com/watch?v=5yOw2_x1Pxg) ] `03-01-2024`. "Top 5 Wireshark tricks to troubleshoot SLOW networks".

JimKurose
* [ [y](https://www.youtube.com/watch?v=2rV4tJkP-CQ) ] `01-15-2022`. "5.4 BGP: the Border Gateway Protocol".
* [ [y](https://www.youtube.com/watch?v=sdIpLuIR1ZY) ] `01-15-2022`. "4.3 The Internet Protocol, part 1".
* [ [y](https://www.youtube.com/watch?v=H9DivCEFuUg) ] `01-15-2022`. "4.3 The Internet Protocol, part 2".
* [ [y](https://www.youtube.com/watch?v=zfzzz9KQOog) ] `01-15-2022`. "4.2 What's inside a router? Part 2."
* [ [y](https://www.youtube.com/watch?v=NfHR94JAO-s) ] `01-15-2022`. "4.2 What's inside a router? Part 1.".
* [ [y](https://www.youtube.com/watch?v=olbr3WZwrdU) ] `01-15-2022`. "4.1 Introduction to the Network Layer".
* [ [y](https://www.youtube.com/watch?v=wEWiYLb7pxg) ] `09-05-2012`. "Traceroute demonstration".

Kevin Du
* [ [y](https://www.youtube.com/watch?v=lfbO-eHt6qg&list=PLwCoMLt7WGjan54CuqeYGnJuMqA-RzQwD) ] `07-31-2021`. "SEED Internet Emulator: Overview".

Linus Tech Tips
* [ [y](https://www.youtube.com/watch?v=_IzyJTcnPu8) ] `11-08-2022`. "Your Router Sucks. Build Your Own Instead!".

My Lesson
* [ [y](https://www.youtube.com/watch?v=BNbPsiCGQzw) ] `01-31-2023`. "IBM IT Support - Complete Course | IT Support Technician - Full Course".

More
* [ [y](https://www.youtube.com/watch?v=_b4dXKB8Pt8) ] `10-23-2019`. danscourses. "The Data Link Layer, MAC Addressing, and the Ethernet Frame".
* [ [y](https://www.youtube.com/watch?v=u7J1YRo1aPY) ] `10-12-2023`. The Bearded I.T. Dad. "Beginner to Pro: A Roadmap for Becoming A Network Engineer".
* [ [y](https://www.youtube.com/watch?v=E5bSumTAHZE) ] `08-30-2021`. Practical Networking. "Network Protocols - ARP, FTP, SMTP, HTTP, SSL, TLS, HTTPS, DNS, DHCP - Networking Fundamentals - L6".

---

## Texts

* Chou, Eric. (2020). _Mastering Python Networking_. 3rd Ed. Packt.
* [ [g](https://github.com/mpcollins/nstda_v2_examples) ][ [g](https://github.com/mpcollins/nsda_examples) ] Collins, Michael. (2017). _Network Security Through Data Analysis: From Data to Action_. 2nd Ed. O'Reilly.
* Edelman, Jason; Scott S. Lowe; & Matt Oswalt. (2018). _Network Programmability and Automation: Skills for the Next-Generation Network Engineer_. O'Reilly.
* Forshaw, James. (2017). _Attacking Network Protocols: A Hacker's Guide to Capture, Analysis, and Exploitation_. [No Starch Press](https://nostarch.com/networkprotocols).
* Gilman, Evan & Doug Barth. (2017). _Zero Trust Networks: Building Secure Systems in Untrusted Networks_. O'Reilly.
* Keshav, Srinivasan. (2012). _Mathematical Foundations of Computer Networking_. Addison-Wesley Professional.
* Kurose, James & Keith Ross. (2016). _Computer Networking: A Top-Down Approach, 7th Ed_.
* Liu, Cricket & Paul Albitz. _DNS and Bind: Help for System Administrators_. 5th Ed. O'Reilly.
* [ [h](https://www.hackers-arise.com/networks-basics) ] OccupyTheWeb. _Network Basics for Hackers_.
* [ [h](https://tcpcc.systemsapproach.org/) ][ [g](https://github.com/SystemsApproach/tcpcc) ] Peterson, Larry L.; Lawrence Brakmo; & Bruce S. Davie. (2022). _TCP Congestion Control: A Systems Approach_. Systems Approach.
* [ [h](https://sdn.systemsapproach.org/) ][ [g](https://github.com/SystemsApproach/sdn) ] Peterson, Larry L. et al. (2021). _Software-Defined Networks: A Systems Approach_. Systems Approach.
* [ h ][ [g](https://github.com/SystemsApproach/book) ] Peterson, Larry L. & Bruce S. Davie. (2021). _Computer Networks: A Systems Approach_. 6e. Morgan Kaufmann Series in Networking.
* Sanders, Chris. (2017). _Practical Packet Analysis: Using Wireshark to Solve Real-World Network Problems_. 3e. [No Starch Press](https://nostarch.com/packetanalysis3).
* Tanenbaum, Andrew & David Wetherall. (2010). _Computer Networks, 5th Ed_. Prentice Hall.
* White, Russ & Ethan Banks. (2017). _Computer Networking Problems and Solutions: An Innovative Approach to Building Resilient, Modern Networks_. Pearson.

* Gast, Matthew. (2013). _802.11ac: A Survival Guide_. O'Reilly.
* Gast, Matthew. (2005). _802.11 Wireless Networks: The Definitive Guide_. 2e. O'Reilly.
* Gulasekaran, Susinder R. & Sundar G. Sankaran. (2021). _Wi-Fi 6 Protocol and Network_. Artech House Publishers.
* Minella, Jennifer. (2022). _Wireless Security Architecture: Designing and Maintaining Secure Wireless for Enterprise_. Wiley.

---

## Figures

* [ [w](https://en.wikipedia.org/wiki/Paul_Baran) ] `1926-2011` Baran, Paul
* [ [w](https://en.wikipedia.org/wiki/Vint_Cerf) ] `1943-----` Cerf, Vint
* [ [w](https://en.wikipedia.org/wiki/David_D._Clark) ] `1944-----` Clark, David
* [ [w](https://en.wikipedia.org/wiki/Donald_Davies) ] `1924-2000` Davies, Donald
* [ [w](https://en.wikipedia.org/wiki/Bob_Kahn) ] `1938-----` Kahn, Bob
* [ [w](https://en.wikipedia.org/wiki/Louis_Pouzin) ] `1931-----` Pouzin, Louis
* [ [w](https://en.wikipedia.org/wiki/David_P._Reed) ] `1952-----` Reed, David
* [ [w](https://en.wikipedia.org/wiki/Jerry_Saltzer) ] `1939-----` Saltzer, Jerry

---

## Terms

* [ [w](https://en.wikipedia.org/wiki/0.0.0.0) ] 0.0.0.0
* [ [w](https://en.wikipedia.org/wiki/6to4) ] 6to4
* [ [w](https://en.wikipedia.org/wiki/Acknowledgement_(data_networks)) ] Acknowledgement (ACK)
* [ [w](https://en.wikipedia.org/wiki/Address_space) ] Address Space
* [ [w](https://en.wikipedia.org/wiki/Asynchronous_Transfer_Mode) ] Asynchronous Transfer Mode
* [ [w](https://en.wikipedia.org/wiki/Australian_Overland_Telegraph_Line) ] Australian Overland Telegraph Line
* [ [w](https://en.wikipedia.org/wiki/Automatic_repeat_request) ] Automatic Repeat Query (ARQ)
* [ [w](https://en.wikipedia.org/wiki/Automatic_repeat_request) ] Automatic Repeat Request (ARQ)
* [ [w](https://en.wikipedia.org/wiki/Autonomous_system_(Internet)) ] Autonomous System (AS)
* [ [w](https://en.wikipedia.org/wiki/Average_per-bit_delivery_cost) ] Average Per-Bit Delivery Cost (APBDC)
* [ [w](https://en.wikipedia.org/wiki/Backbone_network) ] Backbone Network
* [ [w](https://en.wikipedia.org/wiki/Bandwidth_(computing)) ] Bandwidth
* [ [w](https://en.wikipedia.org/wiki/Bandwidth_throttling) ] Bandwidth Throttling
* [ [w](https://en.wikipedia.org/wiki/Bandwidth-delay_product) ] Bandwidth-Delay Product
* [ [w](https://en.wikipedia.org/wiki/Bit_rate) ] Bit Rate
* [ [w](https://en.wikipedia.org/wiki/Bit_time) ] Bit Time
* [ [w](https://en.wikipedia.org/wiki/Bitstream) ] Bitstream
* [ [w](https://en.wikipedia.org/wiki/Border_Gateway_Protocol) ] Border Gateway Protocol (BGP)
* [ [w](https://en.wikipedia.org/wiki/Network_bridge) ] Bridge
* [ [w](https://en.wikipedia.org/wiki/Broadband) ] Broadband
* [ [w](https://en.wikipedia.org/wiki/Broadband_over_power_lines) ] Broadband over Power Lines (BPL)
* [ [w](https://en.wikipedia.org/wiki/Broadcast_address) ] Broadcast Address
* [ [w](https://en.wikipedia.org/wiki/Burstable_billing) ] Burstable Billing
* [ [w](https://en.wikipedia.org/wiki/Cable_layer) ] Cable Ship
* [ [w](https://en.wikipedia.org/wiki/Channel_capacity) ] Channel Capacity
* [ [w](https://en.wikipedia.org/wiki/Circuit_switching) ] Circuit Switching
* [ [w](https://en.wikipedia.org/wiki/Class_of_service) ] Class of Service
* [ [w](https://en.wikipedia.org/wiki/Classful_network) ] Classful Network
* [ [w](https://en.wikipedia.org/wiki/Classless_Inter-Domain_Routing) ] Classless Inter-Domain Routing (CIDR)
* [ [w](https://en.wikipedia.org/wiki/Clock_recovery) ] Clock Recovery
* [ [w](https://en.wikipedia.org/wiki/Coaxial) ] Coaxial
* [ [w](https://en.wikipedia.org/wiki/Coaxial_cable) ] Coaxial Cable
* [ [w](https://en.wikipedia.org/wiki/Committed_information_rate) ] Committed Information Rate
* [ [w](https://en.wikipedia.org/wiki/Communication_protocol) ] Communication Protocol
* [ [w](https://en.wikipedia.org/wiki/Complex_network) ] Complex Network
* [ [w](https://en.wikipedia.org/wiki/Connection-oriented_communication) ] Connection-Oriented Communication
* [ [w](https://en.wikipedia.org/wiki/Connectionless_communication) ] Connectionless Communication
* [ [w](https://en.wikipedia.org/wiki/Constant_envelope) ] Constant Envelope
* [ [w](https://en.wikipedia.org/wiki/Constellation_diagram) ] Constellation Diagram
* [ [w](https://en.wikipedia.org/wiki/Control_plane) ] Control Plane
* [ [w](https://en.wikipedia.org/wiki/Core_router) ] Core Router
* [ [w](https://en.wikipedia.org/wiki/Crosstalk) ] Crosstalk
* [ [w](https://en.wikipedia.org/wiki/Customer-premises_equipment) ] Customer Premises Equipment (CPE)
* [ [w](https://en.wikipedia.org/wiki/Data_cap) ] Data Cap
* [ [w](https://en.wikipedia.org/wiki/Data_circuit-terminating_equipment) ] Data Circuit-Terminating Equipment (DCE)
* [ [w](https://en.wikipedia.org/wiki/Data_link_layer) ] Data Link Layer
* [ [w](https://en.wikipedia.org/wiki/Forwarding_plane) ] Data Plane
* [ [w](https://en.wikipedia.org/wiki/Data_stream) ] Data Stream
* [ [w](https://en.wikipedia.org/wiki/Data_terminal_equipment) ] Data Terminal Equipment (DTE)
* [ [w](https://en.wikipedia.org/wiki/Datagram) ] Datagram
* [ [w](https://en.wikipedia.org/wiki/Deep_packet_inspection) ] Deep Packet Inspection (DPI)
* [ [w](https://en.wikipedia.org/wiki/Default_gateway) ] Default Gateway
* [ [w](https://en.wikipedia.org/wiki/Default_route) ] Default Route
* [ [w](https://en.wikipedia.org/wiki/Delay-tolerant_networking) ] Delay-Tolerant Networking (DTN)
* [ [w](https://en.wikipedia.org/wiki/Distance-vector_routing_protocol) ] Distance-Vector Routing Protocol
* [ [w](https://en.wikipedia.org/wiki/Distribution_frame) ] Distribution Frame
* [ [w](https://en.wikipedia.org/wiki/Diversity_scheme) ] Diversity Scheme
* [ [w](https://en.wikipedia.org/wiki/Routing_in_delay-tolerant_networking) ] DTN Routing
* [ [w](https://en.wikipedia.org/wiki/Dynamic_Host_Configuration_Protocol) ] Dynamic Host Configuration Protocol (DHCP)
* [ [w](https://en.wikipedia.org/wiki/Dynamic_network_analysis) ] Dynamic Network Analysis
* [ [w](https://en.wikipedia.org/wiki/Dynamic_routing) ] Dynamic Routing
* [ [w](https://en.wikipedia.org/wiki/Electrical_cable) ] Electrical Cable
* [ [w](https://en.wikipedia.org/wiki/Electromagnetic_interference) ] Electromagnetic Interference (EMI)
* [ [w](https://en.wikipedia.org/wiki/Encapsulation_(networking)) ] Encapsulation
* [ [w](https://en.wikipedia.org/wiki/End_system) ] End System
* [ [w](https://en.wikipedia.org/wiki/End-to-end_encryption) ] End-to-End Encryption (E2EE)
* [ [w](https://en.wikipedia.org/wiki/End-to-end_principle) ] End-to-End Principle
* [ [w](https://en.wikipedia.org/wiki/Ethernet) ] Ethernet
* [ [w](https://en.wikipedia.org/wiki/Ethernet_hub) ] Ethernet Hub
* [ [w](https://en.wikipedia.org/wiki/Exterior_gateway_protocol) ] Exterior Gateway Protocol
* [ [w](https://en.wikipedia.org/wiki/Failover) ] Failover
* [ [w](https://en.wikipedia.org/wiki/Fiber-optic_cable) ] Fiber Optic Cable
* [ [w](https://en.wikipedia.org/wiki/Flow_control_(data)) ] Flow Control
* [ [w](https://en.wikipedia.org/wiki/Forwarding_plane) ] Forwarding Plane
* [ [w](https://en.wikipedia.org/wiki/Forwarding_information_base) ] Forwarding Table
* [ [w](https://en.wikipedia.org/wiki/Frame_(networking)) ] Frame
* [ [w](https://en.wikipedia.org/wiki/Frame_check_sequence) ] Frame Check Sequence (FCS)
* [ [w](https://en.wikipedia.org/wiki/Frame_Relay) ] Frame Relay
* [ [w](https://en.wikipedia.org/wiki/Gateway_(telecommunications)) ] Gateway
* [ [w](https://en.wikipedia.org/wiki/Goodput) ] Goodput
* [ [w](https://en.wikipedia.org/wiki/Gossip_protocol) ] Gossip Protocol
* [ [w](https://en.wikipedia.org/wiki/Great-circle_navigation) ] Great-Circle Navigation
* [ [w](https://en.wikipedia.org/wiki/Guard_interval) ] Guard Interval
* [ [w](https://en.wikipedia.org/wiki/Handshake_(computing)) ] Handshake
* [ [w](https://en.wikipedia.org/wiki/Header_(computing)) ] Header
* [ [w](https://en.wikipedia.org/wiki/Heterogeneous_network) ] Heterogeneous Network
* [ [w](https://en.wikipedia.org/wiki/Hex_dump) ] Hex Dump
* [ [w](https://en.wikipedia.org/wiki/High_Performance_Computing_Act_of_1991) ] High Performance Computing Act of 1991
* [ [w](https://en.wikipedia.org/wiki/Holddown) ] Holddown
* [ [w](https://en.wikipedia.org/wiki/Hop_(networking)) ] Hop
* [ [w](https://en.wikipedia.org/wiki/Hosts_(file)) ] Hosts File
* [ [w](https://en.wikipedia.org/wiki/Hub_(network_science)) ] Hub
* [ [w](https://en.wikipedia.org/wiki/Human-readable_medium_and_data) ] Human-Readable
* [ [w](https://en.wikipedia.org/wiki/HTTP) ] Hypertext Transfer Protocol (HTTP)
* [ [w](https://en.wikipedia.org/wiki/IEEE_1901) ] IEEE 1901
* [ [w](https://en.wikipedia.org/wiki/IEEE_802) ] IEEE 802
* [ [w](https://en.wikipedia.org/wiki/IEEE_802.1) ] IEEE 802.1
* [ [w](https://en.wikipedia.org/wiki/IEEE_802.11) ] IEEE 802.11
* [ [w](https://en.wikipedia.org/wiki/IEEE_802.1X) ] IEEE 802.1X
* [ [w](https://en.wikipedia.org/wiki/IEEE_802.3) ] IEEE 802.3
* [ [w](https://en.wikipedia.org/wiki/IGMP_snooping) ] IGMP Snooping
* [ [w](https://en.wikipedia.org/wiki/Information_superhighway) ] Information Superhighway
* [ [w](https://en.wikipedia.org/wiki/Intelligent_transportation_system) ] Intelligent Transportation System (ITS)
* [ [w](https://en.wikipedia.org/wiki/Interpacket_gap) ] Inter Packet Gap (IPG)
* [ [w](https://en.wikipedia.org/wiki/Interplanetary_Internet) ] Interplanetary Internet
* [ [w](https://en.wikipedia.org/wiki/Interior_gateway_protocol) ] Interior Gateway Protocol (IGP)
* [ [w](https://en.wikipedia.org/wiki/Internet) ] Internet
* [ [w](https://en.wikipedia.org/wiki/Internet_Assigned_Numbers_Authority) ] Internet Assigned Numbers Authority (IANA)
* [ [w](https://en.wikipedia.org/wiki/Internet_backbone) ] Internet Backbone
* [ [w](https://en.wikipedia.org/wiki/Internet_checksum) ] Internet Checksum
* [ [w](https://en.wikipedia.org/wiki/Internet_exchange_point) ] Internet Exchange Point (IXP)
* [ [w](https://en.wikipedia.org/wiki/Internet_Group_Management_Protocol) ] Internet Group Management Protocol (IGMP)
* [ [w](https://en.wikipedia.org/wiki/History_of_the_Internet) ] Internet History
* [ [w](https://en.wikipedia.org/wiki/Internet_of_things) ] Internet of Things (IoT)
* [ [w](https://en.wikipedia.org/wiki/Internet_of_vehicles) ] Internet of Vehicles (IoV)
* [ [w](https://en.wikipedia.org/wiki/Internet_Protocol) ] Internet Protocol (IP)
* [ [w](https://en.wikipedia.org/wiki/Internet_protocol_suite) ] Internet Protocol Suite (TCP/IP)
* [ [w](https://en.wikipedia.org/wiki/Internet_Protocol_version_4) ] Internet Protocol version 4 (IPv4)
* [ [w](https://en.wikipedia.org/wiki/IPv6) ] Internet Protocol version 6 (IPv6)
* [ [w](https://en.wikipedia.org/wiki/IP_address) ] IP Address
* [ [w](https://en.wikipedia.org/wiki/IP_multicast) ] IP Multicast
* [ [w](https://en.wikipedia.org/wiki/IP_tunnel) ] IP Tunnel
* [ [w](https://en.wikipedia.org/wiki/IPv4_address_exhaustion) ] IPv4 Address Exhaustion
* [ [w](https://en.wikipedia.org/wiki/IPv4_shared_address_space) ] IPv4 Shared Address Space
* [ [w](https://en.wikipedia.org/wiki/Kerberos_(protocol)) ] Kerberos
* [ [w](https://en.wikipedia.org/wiki/Leased_line) ] Leased Line
* [ [w](https://en.wikipedia.org/wiki/Link-local_address) ] Link-Local Address
* [ [w](https://en.wikipedia.org/wiki/Link_aggregation) ] Link Aggregation
* [ w ] Link Aggregation Control Protocol (LACP)
* [ [w](https://en.wikipedia.org/wiki/Link_encryption) ] Link Encryption
* [ [w](https://en.wikipedia.org/wiki/Link_layer) ] Link Layer
* [ [w](https://en.wikipedia.org/wiki/Link-state_routing_protocol) ] Link-State Routing Protocol
* [ [w](https://en.wikipedia.org/wiki/Load_(computing)) ] Load
* [ [w](https://en.wikipedia.org/wiki/Load_balancing_(computing)) ] Load Balancing
* [ [w](https://en.wikipedia.org/wiki/Local_area_network) ] Local Area Network (LAN)
* [ [w](https://en.wikipedia.org/wiki/Localhost) ] localhost
* [ [w](https://en.wikipedia.org/wiki/Longest_prefix_match) ] Longest Prefix Match
* [ [w](https://en.wikipedia.org/wiki/Loopback) ] Loopback
* [ [w](https://en.wikipedia.org/wiki/Maximal-ratio_combining) ] Maximal-Ratio Combining (MRC)
* [ [w](https://en.wikipedia.org/wiki/Maximum_transmission_unit) ] Maximum Transmission Unit (MTU)
* [ [w](https://en.wikipedia.org/wiki/Medium_access_control) ] Media Access Control (MAC)
* [ [w](https://en.wikipedia.org/wiki/MAC_address) ] Media Access Control (MAC) Address
* [ [w](https://en.wikipedia.org/wiki/Message_switching) ] Message Switching
* [ [w](https://en.wikipedia.org/wiki/Metcalfe%27s_law) ] Metcalfe's Law
* [ [w](https://en.wikipedia.org/wiki/Wireless_ad_hoc_network) ] Mobile Ad Hoc Network (MANET)
* [ [w](https://en.wikipedia.org/wiki/Modulation) ] Modulation
* [ [w](https://en.wikipedia.org/wiki/Modem) ] Modulator-Demodulator (Modem)
* [ [w](https://en.wikipedia.org/wiki/Multicast_address) ] Multicast Address
* [ [w](https://en.wikipedia.org/wiki/National_Information_Infrastructure) ] National Information Infrastructure (NII)
* [ [w](https://en.wikipedia.org/wiki/National_Science_Foundation_Network) ] National Science Foundation Network (NSFNET)
* [ [w](https://en.wikipedia.org/wiki/Net_bias) ] Net Bias
* [ [w](https://en.wikipedia.org/wiki/Network_address_translation) ] Network Address Translation (NAT)
* [ [w](https://en.wikipedia.org/wiki/Network_administrator) ] Network Administrator
* [ [w](https://en.wikipedia.org/wiki/Network_architecture) ] Network Architecture
* [ [w](https://en.wikipedia.org/wiki/Network_congestion) ] Network Congestion
* [ [w](https://en.wikipedia.org/wiki/Network_delay) ] Network Delay
* [ [w](https://en.wikipedia.org/wiki/Network_effect) ] Network Effect
* [ [w](https://en.wikipedia.org/wiki/Networking_hardware) ] Network Equipment
* [ [w](https://en.wikipedia.org/wiki/Network_forensics) ] Network Forensics
* [ [w](https://en.wikipedia.org/wiki/Network_interface_controller) ] Network Interface Controller (NIC)
* [ [w](https://en.wikipedia.org/wiki/Network_layer) ] Network Layer
* [ [w](https://en.wikipedia.org/wiki/Metrics_(networking)) ] Network Metric
* [ [w](https://en.wikipedia.org/wiki/Network_packet) ] Network Packet
* [ [w](https://en.wikipedia.org/wiki/Network_planning_and_design) ] Network Planning and Design
* [ [w](https://en.wikipedia.org/wiki/Network_science) ] Network Science
* [ [w](https://en.wikipedia.org/wiki/Network_security) ] Network Security
* [ [w](https://en.wikipedia.org/wiki/Network_Security_Services) ] Network Security Services (NSS)
* [ [w](https://en.wikipedia.org/wiki/Network_segment) ] Network Segment
* [ [w](https://en.wikipedia.org/wiki/Network_service_provider) ] Network Service Provider (NSP)
* [ [w](https://en.wikipedia.org/wiki/Network_socket) ] Network Socket
* [ [w](https://en.wikipedia.org/wiki/Network_theory) ] Network Theory
* [ [w](https://en.wikipedia.org/wiki/Network_topology) ] Network Topology
* [ [w](https://en.wikipedia.org/wiki/Network_traffic) ] Network Traffic
* [ [w](https://en.wikipedia.org/wiki/Next-generation_network) ] Next-Generation Network (NGN)
* [ [w](https://en.wikipedia.org/wiki/Node_(networking)) ] Node
* [ [w](https://en.wikipedia.org/wiki/Open_Shortest_Path_First) ] Open Shortest Path First (OSPF)
* [ [w](https://en.wikipedia.org/wiki/OSI_model) ] Open Systems Interconnection (OSI) Model
* [ [w](https://en.wikipedia.org/wiki/Optical_fiber) ] Optical Fiber
* [ [w](https://en.wikipedia.org/wiki/Out-of-order_delivery) ] Out-of-Order Delivery
* [ [w](https://en.wikipedia.org/wiki/Overhead_(computing)) ] Overhead
* [ [w](https://en.wikipedia.org/wiki/Overlay_network) ] Overlay Network
* [ [w](https://en.wikipedia.org/wiki/Network_packet) ] Packet
* [ [w](https://en.wikipedia.org/wiki/Packet_analyzer) ] Packet Analysis
* [ [w](https://en.wikipedia.org/wiki/Packet_capture_appliance) ] Packet Capture Appliance
* [ [w](https://en.wikipedia.org/wiki/Packet_forwarding) ] Packet Forwarding
* [ [w](https://en.wikipedia.org/wiki/Packet_loss) ] Packet Loss
* [ [w](https://en.wikipedia.org/wiki/Packet_switching) ] Packet Switching
* [ [w](https://en.wikipedia.org/wiki/Patch_panel) ] Patch Panel
* [ [w](https://en.wikipedia.org/wiki/Peak_information_rate) ] Peak Information Rate (PIR)
* [ [w](https://en.wikipedia.org/wiki/Peering) ] Peering
* [ [w](https://en.wikipedia.org/wiki/PHY-Level_Collision_Avoidance) ] PHY-Level Collision Avoidance (PLCA)
* [ [w](https://en.wikipedia.org/wiki/Physical_layer) ] Physical Layer
* [ [w](https://en.wikipedia.org/wiki/Ping_of_death) ] Ping of Death
* [ [w](https://en.wikipedia.org/wiki/Point-to-Point_Protocol) ] Point-to-Point Protocol
* [ [w](https://en.wikipedia.org/wiki/Port_(computer_networking)) ] Port
* [ [w](https://en.wikipedia.org/wiki/Power-line_communication) ] Power-Line Communication (PLC)
* [ [w](https://en.wikipedia.org/wiki/Syncword) ] Preamble
* [ [w](https://en.wikipedia.org/wiki/Preboot_Execution_Environment) ] Preboot Execution Environment (PXE)
* [ [w](https://en.wikipedia.org/wiki/Private_network) ] Private Network
* [ [w](https://en.wikipedia.org/wiki/Processing_delay) ] Processing Delay
* [ [w](https://en.wikipedia.org/wiki/Signal_propagation_delay) ] Propagation Delay
* [ [w](https://en.wikipedia.org/wiki/Proprietary_protocol) ] Proprietary Protocol
* [ [w](https://en.wikipedia.org/wiki/Protocol_analyzer) ] Protocol Analysis
* [ [w](https://en.wikipedia.org/wiki/Protocol_data_unit) ] Protocol Data Unit
* [ [w](https://en.wikipedia.org/wiki/Protocol_ossification) ] Protocol Ossification
* [ [w](https://en.wikipedia.org/wiki/Open_Shortest_Path_First) ] Open Shortest Path First (OSPF)
* [ [w](https://en.wikipedia.org/wiki/Queuing_delay) ] Queueing Delay
* [ [w](https://en.wikipedia.org/wiki/Queueing_theory) ] Queueing Theory
* [ [w](https://en.wikipedia.org/wiki/QUIC) ] QUIC
* [ [w](https://en.wikipedia.org/wiki/Electromagnetic_interference) ] Radio-Frequency Interference (RFI)
* [ [w](https://en.wikipedia.org/wiki/Real-time_communication) ] Real-Time Communication (RTC)
* [ [w](https://en.wikipedia.org/wiki/Real-time_computing) ] Real-Time Computing
* [ [w](https://en.wikipedia.org/wiki/Redundancy_(engineering)) ] Redundancy
* [ [w](https://en.wikipedia.org/wiki/Reed%27s_law) ] Reed's Law
* [ [w](https://en.wikipedia.org/wiki/Reliability_(computer_networking)) ] Reliability
* [ [w](https://en.wikipedia.org/wiki/Reliable_byte_stream) ] Reliable Byte Stream
* [ [w](https://en.wikipedia.org/wiki/Reliable_multicast) ] Reliable Multicast
* [ [w](https://en.wikipedia.org/wiki/Reserved_IP_addresses) ] Reserved IP Address
* [ [w](https://en.wikipedia.org/wiki/Residential_gateway) ] Residential Gateway
* [ [w](https://en.wikipedia.org/wiki/Reverse_proxy) ] Reverse Proxy
* [ [w](https://en.wikipedia.org/wiki/Round-robin_scheduling) ] Round-Robin Scheduling
* [ [w](https://en.wikipedia.org/wiki/Round-trip_delay) ] Round-Trip Delay (RTD)
* [ [w](https://en.wikipedia.org/wiki/Round-trip_delay) ] Round-Trip Time (RTT)
* [ [w](https://en.wikipedia.org/wiki/Route_(command)) ] `route`
* [ [w](https://en.wikipedia.org/wiki/Route_poisoning) ] Route Poisoning
* [ [w](https://en.wikipedia.org/wiki/Router_(computing)) ] Router
* [ [w](https://en.wikipedia.org/wiki/Routing) ] Routing
* [ [w](https://en.wikipedia.org/wiki/Convergence_(routing)) ] Routing Convergence
* [ [w](https://en.wikipedia.org/wiki/Routing_Information_Protocol) ] Routing Information Protocol (RIP)
* [ [w](https://en.wikipedia.org/wiki/Routing_loop) ] Routing Loop
* [ [w](https://en.wikipedia.org/wiki/Routing_protocol) ] Routing Protocol
* [ [w](https://en.wikipedia.org/wiki/Routing_table) ] Routing Table
* [ [w](https://en.wikipedia.org/wiki/Secure_Network_Programming) ] Secure Network Programming
* [ [w](https://en.wikipedia.org/wiki/Session_(computer_science)) ] Session
* [ [w](https://en.wikipedia.org/wiki/Sink_(computing)) ] Sink
* [ [w](https://en.wikipedia.org/wiki/Slashdot_effect) ] Slashdot Effect
* [ [w](https://en.wikipedia.org/wiki/Software-defined_networking) ] Software-Defined Networking (SDN)
* [ [w](https://en.wikipedia.org/wiki/SD-WAN) ] Software-Defined Wide Area Network (SD-WAN)
* [ [w](https://en.wikipedia.org/wiki/Split_horizon_route_advertisement) ] Split-Horizon Route Advertisement
* [ [w](https://en.wikipedia.org/wiki/Spoofing_attack) ] Spoofing Attack
* [ [w](https://en.wikipedia.org/wiki/Static_routing) ] Static Routing
* [ [w](https://en.wikipedia.org/wiki/Statistical_time-division_multiplexing) ] Statistical Multiplexing
* [ [w](https://en.wikipedia.org/wiki/Store_and_forward) ] Store and Forward
* [ [w](https://en.wikipedia.org/wiki/Submarine_branching_unit) ] Submarine Branchine Unit
* [ [w](https://en.wikipedia.org/wiki/Submarine_communications_cable) ] Submarine Communications Cable
* [ [w](https://en.wikipedia.org/wiki/List_of_international_submarine_communications_cables) ] submarine communications cables
* [ [w](https://en.wikipedia.org/wiki/Subnet) ] Subnet
* [ [w](https://en.wikipedia.org/wiki/Network_switch) ] Switch
* [ [w](https://en.wikipedia.org/wiki/Syncword) ] Syncword
* [ [w](https://en.wikipedia.org/wiki/TAT-8) ] TAT-8
* [ [w](https://en.wikipedia.org/wiki/Telecommunication_circuit) ] Telecommunications Circuit
* [ [w](https://en.wikipedia.org/wiki/Telecommunications_network) ] Telecommunications Network
* [ [w](https://en.wikipedia.org/wiki/Telegrapher%27s_equations) ] Telegrapher's Equations
* [ [w](https://en.wikipedia.org/wiki/Telephone_network) ] Telephone Network
* [ [w](https://en.wikipedia.org/wiki/Network_throughput) ] Throughput
* [ [w](https://en.wikipedia.org/wiki/Tiered_Internet_service) ] Tiered Service
* [ [w](https://en.wikipedia.org/wiki/Transatlantic_communications_cable) ] Trans Atlantic Telecommunications Cable
* [ [w](https://en.wikipedia.org/wiki/Transatlantic_telegraph_cable) ] Trans Atlantic Telegraph Cable
* [ [w](https://en.wikipedia.org/wiki/Transmission_Control_Protocol) ] Transmission Control Protocol (TCP)
* [ [w](https://en.wikipedia.org/wiki/Transmission_delay) ] Transmission Delay
* [ [w](https://en.wikipedia.org/wiki/Transmit_diversity) ] Transmit Diversity
* [ [w](https://en.wikipedia.org/wiki/Transport_layer) ] Transport Layer
* [ [w](https://en.wikipedia.org/wiki/Trunking) ] Trunking
* [ [w](https://en.wikipedia.org/wiki/Tunneling_protocol) ] Tunneling Protocol
* [ [w](https://en.wikipedia.org/wiki/Twisted_pair) ] Twisted Pair
* [ [w](https://en.wikipedia.org/wiki/User_Datagram_Protocol) ] User Datagram Protocol (UDP)
* [ [w](https://en.wikipedia.org/wiki/Vehicular_ad_hoc_network) ] Vehicular Ad Hoc Network (VANET)
* [ [w](https://en.wikipedia.org/wiki/Virtual_circuit) ] Virtual Circuit
* [ [w](https://en.wikipedia.org/wiki/Virtual_Link_Trunking) ] Virtual Link Trunking (VLT)
* [ [w](https://en.wikipedia.org/wiki/Wide_area_network) ] Wide Area Network (WAN)
* [ [w](https://en.wikipedia.org/wiki/Wireless_ad_hoc_network) ] Wireless Ad Hoc Network (WANET)

* [ [w](https://en.m.wikipedia.org/wiki/Interactive_Connectivity_Establishment) ] Interactive Connectivity Establishment (ICE)
* [ [w](https://en.m.wikipedia.org/wiki/Network_address_translation) ] Network Address Translation (NAT)
  * RFC 1631
  * RFC 2663
  * RFC 4787
* [ [w](https://en.m.wikipedia.org/wiki/QUIC) ] Quick UDP Internet Connections (QUIC)
* [ [w](https://en.wikipedia.org/wiki/Real-time_Transport_Protocol) ] Real-Time Transport Protocol (RTP)
* [ [w](https://en.wikipedia.org/wiki/RTP_Control_Protocol) ] RTP Control Protocol (RTCP)
  * RFC 3550
* [ [w](https://en.wikipedia.org/wiki/Secure_Real-time_Transport_Protocol) ] Secure Real-Time Transport Protocol (SRTP)
* [ [w](https://en.m.wikipedia.org/wiki/Session_Initiation_Protocol) ] Session Initiation Protocol (SIP)
* [ [w](https://en.m.wikipedia.org/wiki/STUN) ] Session Traversal Utilities for NAT (STUN)
  * RFC 3489
  * RFC 5389
  * [ [Apple Platform Security](https://support.apple.com/guide/security/facetime-security-seca331c55cd) ] FaceTime security

* [ w ] Maximum Segment Size (MSS) [ [cloudflare](https://www.cloudflare.com/learning/network-layer/what-is-mss/) ]
* [ [w](https://en.wikipedia.org/wiki/Maximum_transmission_unit) ] Maximum Transmission Unit (MTU) [ [cloudflare](https://www.cloudflare.com/learning/network-layer/what-is-mtu/) ]

AXT Alien Crosstalk <br>
ADC Analog-to-Digital Converter [w](https://en.wikipedia.org/wiki/Analog-to-digital_converter) <br>
Bandwidth <br>
BER Bit Error Rate <br>
Bluetooth 5.0, 5.1, 5.2 <br>
BLE Bluetooth Low Energy <br>
Bonjour <br>
BGP Border Gateway Protocol <br>
CMP Communications Multipurpose Plenum <br>
CMR Communications Multipurpose Riser <br>
Crosstalk <br>
CRC Cyclic Redundancy Check <br>
DAC Digital-to-Analog Converter [w](https://en.wikipedia.org/wiki/Digital-to-analog_converter) <br>
DNS Domain Name System <br>
DHCP Dynamic Host Configuration Protocol <br>
DNS Sinkhole [w](https://en.wikipedia.org/wiki/DNS_sinkhole) <br>
EMI Electromagnetic Interference <br>
ELFEXT Equal Level FEXT <br>
Ethernet <br>
FEXT Far-End Crosstalk <br>
FCS Frame Check Sequence <br>
Gain [w](https://en.wikipedia.org/wiki/Gain_(antenna)) <br>
Higher-Order Modulation (1024-QAM) <br>
Hosts File [w](https://en.wikipedia.org/wiki/Hosts_(file)) <br>
IF Intermediate Frequency [w](https://en.wikipedia.org/wiki/Intermediate_frequency) <br>
IP Internet Protocol <br>
ISP Internet Service Provider <br>
LAN Local Area Network (e.g., home or small business) <br>
LTE Long Term Evolution <br>
MCS Modulation and Coding Scheme <br>
MU-MIMO Multi-User Multi-Input Multi-Output <br>
mDNS Multicast DNS <br>
NEXT Near-End Crosstalk [dB per foot] <br>
NFC Near Field Communication <br>
NAS Network-Attached Storage [Synology](https://www.synology.com/en-us/solution/what_is_nas) [w](https://en.wikipedia.org/wiki/Network-attached_storage) <br>
Nyquist Frequency [w](https://en.wikipedia.org/wiki/Nyquist_frequency) <br>
Nyquist-Shannon Sampling Theorem [w](https://en.wikipedia.org/wiki/Nyquist%E2%80%93Shannon_sampling_theorem) <br>
OFDMA Orthogonal Frequency-Division Multiple Access <br>
PCI Peripheral Component Interconnect [w](https://en.wikipedia.org/wiki/Peripheral_Component_Interconnect) <br>
PAN Personal Area Network (e.g., Mobile Hotspot) <br>
Personal Cloud [w](https://en.wikipedia.org/wiki/Personal_cloud) <br>
PVC Polyvinyl Chloride <br>
PSNEXT Power Sum NEXT <br>
QoS Quality of Service <br>
RF Radio Frequency <br>
RSSI Received Signal Strength Indicator [w](https://en.wikipedia.org/wiki/Received_signal_strength_indication) <br>
Receiver <br>
RFID <br>
RJ Registered Jack <br>
RPS Redundant Power Supply <br>
Sampling [w](https://en.wikipedia.org/wiki/Sampling_(signal_processing)) <br>
Shannon-Hartley Theorem [w](https://en.wikipedia.org/wiki/Shannon%E2%80%93Hartley_theorem) <br>
STP Shielded Twisted Pair <br>
SNR Signal-to-Noise Ratio <br>
SFP Small Form-Factor Pluggable [w](https://en.wikipedia.org/wiki/Small_form-factor_pluggable_transceiver) <br>
SDN Software-Defined Networking <br>
SAN Storage-Area Network [w](https://en.wikipedia.org/wiki/Storage_area_network) <br>
TCP/IP (IP Suite) <br>
Throughput <br>
Transceiver <br>
TCP Transmission Control Protocol <br>
Transmitter <br>
TWT Target Wake Time <br>
Ultra Wideband <br>
UPS Uninterruptible Power Supply [w](https://en.wikipedia.org/wiki/Uninterruptible_power_supply) <br>
UPnP Universal Plug and Play <br>
USB Universal Serial Bus [w](https://en.wikipedia.org/wiki/USB) <br>
UDP User Datagram Protocol <br>
UTP Unshielded Twisted Pair <br>
VLAN Virtual LAN <br>
WAN Wide Area Network (e.g., the Internet) <br>
WAN Failover <br>
WiFi Wireless Fidelity <br>
WiFi 6 (802.11ax) <br>
WiFi 6E <br>
WPA WiFi Protected Access <br>
WSN Wireless Sensor Network [w](https://en.wikipedia.org/wiki/Wireless_sensor_network) <br>
Z-Wave <br>
Zero-Configuration Networking <br>
Zigbee <br>

---

## Notes

Hub - a physical layer (layer 1) device that allows for connections from many computers at once
* each device connected to a hub communicates with every other hub-connected device simultaneously, which generates a lot of noise and a collision domain
* each hub-connected device must determine whether incoming traffic is meant for it

Collision Domain - a network segment where only one device can communicate at a time
* if multiple systems attempt to send data at the same time then the electrical pulses sent across the cable can interfere with each other, which slows down network communication

Switch ("switching hub") - a data link layer (layer 2) device which allows for intra-LAN communication
* a switch inspect ethernet data to determine which computer to forward it to
* benefits: reduces/eliminates collision domain; decreases retransmission; increases throughput

Router - a network layer (layer 3) device which allows for inter-LAN communication
* switches connect computers within a single network whereas routers connect computers across networks
* a router inspects IP data to determine where to forward it
* a routing table contains information about how to route traffic

---
